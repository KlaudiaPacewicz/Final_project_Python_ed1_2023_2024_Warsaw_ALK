```python
import pandas as pd
import numpy as np
```

# Import input data

```python
r1 = pd.read_excel('./../../original_input_data/clinical_metadata/Repository1_sample_clinical_metadata.xlsx')
r1
```

## Replace NA with NaN/NaT values

```python
r1 = r1.replace(r'N/A', np.nan, regex=True)
r1
```

## Calculate age at diagnosis

```python
r1['donor_age_at_diagnosis'] = r1['biopsy_year'] - r1['birth_year']
r1
```

## Identification of patients treated with platinum-based chemotherapy or PARPi (PARPi treated patients will be further used for model evaluation)

```python
platinum_drugs = ['Carboplatin', 'Cisplatin', 'Oxaliplatin']

def platinum_treatment(treatment):
    if isinstance(treatment, str):
        if any(drug in treatment for drug in platinum_drugs):
            return 1  
        else:
            return 0 
    elif pd.isna(treatment):
            return 'use_as_blind_test'
    else:
        return 0

# Test ffor platinum_treatment function
# x_test = 'Oxaliplatin'
# result = platinum_treatment(x_test)
# print(result)    

parpi_inhibitors = ['Olaparib', 'Niraparib', 'Rucaparib', 'Talazoparib']

def parpi_treatment(treatment):
    if isinstance(treatment, str):
        if any(drug in treatment for drug in parpi_inhibitors):
            return 1
        else:
            return 0

# Test for parpi_treatment function
            
#y_test = 'jdks'
#result = parpi_treatment(y_test)
#print(result)


# Aplication of platinum_treatment and parpi_treatment functions to r1 dataframe

r1['platinum-based_chemotherapy'] = r1['treatment'].apply(lambda x: platinum_treatment(x))
r1['parpi_treatment'] = r1['treatment'].apply(lambda x: parpi_treatment(x))
r1
```

## Calculation of stable disease period depending on the response measure

```python
def calculate_stable_disease_period_days(x):
    if x['first_response_measure'] in ['SD'] and x['last_response_measure'] == 'SD':
        return (x['last_response_date'] - x['treatment_start_date']).days
    elif x['first_response_measure'] in ['SD'] and pd.isna(x['last_response_measure']):
        return (x['first_response_date'] - x['treatment_start_date']).days
    else:
        return 'Not applicable'


r1["stable_disease_period_days"] = r1.apply(calculate_stable_disease_period_days, axis=1)
r1
```

### Stable disease period - days to month transformation

```python
r1["stable_disease_period_days"] = pd.to_numeric(r1["stable_disease_period_days"], errors='coerce')

r1["stable_disease_period_months"] = round(r1["stable_disease_period_days"] / 30.41)
r1
```

## Calculation of progression-free survival period depending on the response measure

```python
def calculate_progression_free_survival_days(x):
    if x['first_response_measure'] == 'PD':
        return (x['first_response_date'] - x['treatment_start_date']).days
    elif x['first_response_measure'] in ['SD', 'PR', 'CR'] and x['last_response_measure'] == 'PD':
        return (x['last_response_date'] - x['treatment_start_date']).days
    else:
        return 'Not applicable'


r1["progression_free_survival_days"] = r1.apply(calculate_progression_free_survival_days, axis=1)
r1["progression_free_survival_days"] = r1["progression_free_survival_days"].fillna('Not applicable')
r1
```

### Progression-free survival period - days to months tranformation

```python
r1["progression_free_survival_days"] = pd.to_numeric(r1["progression_free_survival_days"], errors='coerce')

r1["progression_free_survival_months"] = round(r1["progression_free_survival_days"] / 30.41)
r1
```

## Labels preparation

```python
def sample_labeling(x):
    if x['platinum-based_chemotherapy'] == 1:
        if (x['first_response_measure'] in ['PR', 'CR'] and 
           (x['last_response_measure'] in ['PR', 'CR', 'SD', 'PD'] or 
            pd.isna(x['last_response_measure']))):
            return 'true_responder'
        elif x['first_response_measure'] == 'SD':
            if x['last_response_measure'] in ['PR', 'CR']:
                return 'possible_reponder'
            elif (x['last_response_measure'] == 'SD' or 
                  pd.isna(x['last_response_measure'])):
                if x['stable_disease_period_months'] >= 6:
                    return 'responder_based_on_SD'
                else:
                    return 'nonresponder_based_on_SD'
            else:
                if x['last_response_measure'] == 'PD':
                    if x['progression_free_survival_months'] >= 6:
                        return 'responder_based_on_PFS'
                    else:
                        return 'nonresponder_based_on_PFS'
        elif x['first_response_measure'] == 'PD':
                return 'true_nonresponder'
        else:
            if pd.isna(x['first_response_measure']):
                return 'other_use_as_test'
    else:
        if x['platinum-based_chemotherapy'] in ['use_as_blind_test']:
            return 'other_use_as_test'

# Test for sample_labeling function

x_test = {
    'platinum-based_chemotherapy': 'use_as_blind_test',
    'first_response_measure': 'PR',
    'last_response_measure':  'PD',
    'stable_disease_period_months': 10,
    'progression_free_survival_months': 3,
}

result = sample_labeling(x_test)
print(result) 


r1['label'] = r1.apply(sample_labeling, axis=1)
r1
```
