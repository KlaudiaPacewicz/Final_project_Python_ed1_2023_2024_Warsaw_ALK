## Features cleaning based on observations above

```python
features_cleaned = features
features_cleaned
```

#### Merge duplication (CNV) up to 30.000bp and leave >30.000bp

```python
features_cleaned['cnv_DUP_len_0_30000'] = features_cleaned['cnv_DUP_len_0_30'] + features_cleaned['cnv_DUP_len_100_1000'] + features_cleaned ['cnv_DUP_len_1000_30000']

# Checking data correctness

features3 = features_cleaned[['cnv_DUP_len_0_30000', 'cnv_DUP_len_0_30', 'cnv_DUP_len_100_1000', 'cnv_DUP_len_1000_30000']]
features3['cnv_DUP_len_0_30000'].plot(kind="hist")

# Removing 'cnv_DUP_len_0_30', 'cnv_DUP_len_100_1000', 'cnv_DUP_len_1000_30000' from features

features_cleaned = features_cleaned.drop('cnv_DUP_len_0_30', axis='columns')
features_cleaned = features_cleaned.drop('cnv_DUP_len_100_1000', axis='columns')
features_cleaned = features_cleaned.drop('cnv_DUP_len_1000_30000', axis='columns')
features_cleaned
```

#### Merge deletion (CNV) up to 100.000bp and leave >100.000bp

```python
features_cleaned['cnv_DEL_len_0_100000'] = features_cleaned['cnv_DEL_len_0_30'] + features_cleaned['cnv_DEL_len_30_3500'] + features_cleaned ['cnv_DEL_len_3500_100000']

# Checking data correctness

features4 = features_cleaned[['cnv_DEL_len_0_100000', 'cnv_DEL_len_0_30', 'cnv_DEL_len_30_3500', 'cnv_DEL_len_3500_100000']]
features4
features4['cnv_DEL_len_0_100000'].plot(kind="hist")

# Removing 'cnv_DEL_len_0_30', 'cnv_DEL_len_30_3500', 'cnv_DEL_len_3500_100000' from features

features_cleaned = features_cleaned.drop('cnv_DEL_len_0_30', axis='columns')
features_cleaned = features_cleaned.drop('cnv_DEL_len_30_3500', axis='columns')
features_cleaned = features_cleaned.drop('cnv_DEL_len_3500_100000', axis='columns')
features_cleaned
```

#### Merge diploid regions (CNV) up to 100.000bp and leave >100.000bp

```python
features_cleaned['cnv_DIPLOID_len_0_100000'] = features_cleaned['cnv_DIPLOID_len_0_30'] + features_cleaned['cnv_DIPLOID_len_30_500'] + features_cleaned ['cnv_DIPLOID_len_500_100000']

# Checking data correctness

features5 = features_cleaned[['cnv_DIPLOID_len_0_100000', 'cnv_DIPLOID_len_0_30', 'cnv_DIPLOID_len_30_500', 'cnv_DIPLOID_len_500_100000']]
features5
features5['cnv_DIPLOID_len_0_100000'].plot(kind="hist")

# Removing 'cnv_DIPLOID_len_0_30', 'cnv_DIPLOID_len_30_500', 'cnv_DIPLOID_len_500_100000' from features

features_cleaned = features_cleaned.drop('cnv_DIPLOID_len_0_30', axis='columns')
features_cleaned = features_cleaned.drop('cnv_DIPLOID_len_30_500', axis='columns')
features_cleaned = features_cleaned.drop('cnv_DIPLOID_len_500_100000', axis='columns')
features_cleaned
```

#### Remove 'sv_n_svclass_variance' feature

```python
features_cleaned = features_cleaned.drop('sv_n_svclass_variance', axis='columns')
features_cleaned
```

#### Merge sv_dup up to 30.000bp and leave >30.000bp

```python
features_cleaned['sv_dup_0_30000'] = features_cleaned['sv_dup_0_30'] + features_cleaned['sv_dup_30-100'] + features_cleaned ['sv_dup_100-1000'] + features_cleaned['sv_dup_1000_30000']

# Checking data correctness

features6 = features_cleaned[['sv_dup_0_30000','sv_dup_0_30', 'sv_dup_30-100', 'sv_dup_100-1000', 'sv_dup_1000_30000']]
features6
features6['sv_dup_0_30000'].plot(kind="hist")

# Removing 'sv_dup_0_30', 'sv_dup_30-100', 'sv_dup_100-1000', 'sv_dup_1000_30000' from features

features_cleaned = features_cleaned.drop('sv_dup_0_30', axis='columns')
features_cleaned = features_cleaned.drop('sv_dup_30-100', axis='columns')
features_cleaned = features_cleaned.drop('sv_dup_100-1000', axis='columns')
features_cleaned = features_cleaned.drop('sv_dup_1000_30000', axis='columns')
features_cleaned
```

#### Merge sv_del up to 100.000bp and leave >100.000bp
